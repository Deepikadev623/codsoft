```python
import pandas as pd
import zipfile
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report
from nltk.corpus import stopwords
from sklearn.utils.class_weight import compute_class_weight

import nltk
nltk.download('stopwords')

with zipfile.ZipFile('train_data.txt.zip', 'r') as z:
    z.extractall()
df = pd.read_csv("train_data.txt", delimiter=':::', header=None, encoding="latin-1", engine='python')
df.columns = ['ID', 'Movie', 'Genre', 'Description']
df['Description'] = df['Description'].fillna('')
df['Genre'] = df['Genre'].fillna('')
X = df['Description']
y = df['Genre']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 2: Vectorize the movie descriptions using TF-IDF
vectorizer = TfidfVectorizer(stop_words=stopwords.words('english'), max_features=5000)
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)

# Step 3: Compute class weights to handle imbalanced classes
class_weights = compute_class_weight('balanced', classes=y.unique(), y=y_train)
class_weight_dict = {i: class_weights[i] for i in range(len(y.unique()))}

# Step 4: Train a model (Naive Bayes in this case)
model = MultinomialNB()
model.fit(X_train_vec, y_train)

# Step 5: Predict genres
y_pred = model.predict(X_test_vec)

# Step 6: Evaluate model
print("Classification Report:\n", classification_report(y_test, y_pred))

# Step 7: Show the output with movie description and predicted genre
output = pd.DataFrame({
    'Movie Description': X_test,
    'Predicted Genre': y_pred
})

# Step 8: Display first 20 predictions
print(output.head(20))  # Display the first 20 predictions
```
